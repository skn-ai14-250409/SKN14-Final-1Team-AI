```python
import pandas as pd

QA_PATH  = "/content/ragas_evaluation_result_QA.csv"
REF_PATH = "/content/ragas_evaluation_result_original.csv"
ID_COL   = "user_input"
METRICS  = ["context_recall", "faithfulness", "factual_correctness(mode=f1)"]
```

```python
# 병합
qa  = pd.read_csv(QA_PATH)
ref = pd.read_csv(REF_PATH)

merged = qa.merge(ref, on=ID_COL, suffixes=("_QA", "_REF"))
print(f"[INFO] 병합된 행 수: {len(merged)}")
```

```python
display(merged)
```

```python
# 차이 계산: QA - REF
for m in METRICS:
    merged[f"delta_{m}"] = merged[f"{m}_QA"] - merged[f"{m}_REF"]
```

```python
# 정량 요약: 각 지표별 QA/REF 평균, 차 평균, 개선 비율
rows = []
for m in METRICS:
    d = merged[f"delta_{m}"]
    rows.append({
        "metric": m,
        "QA_mean": merged[f"{m}_QA"].mean(),
        "REF_mean": merged[f"{m}_REF"].mean(),
        "Delta_mean(QA-REF)": d.mean(),
        "Improved_ratio(Δ>0)": (d > 0).mean()
    })

summary = pd.DataFrame(rows).round(4)
display(summary)
```

```python
# 사례 분석: 지표별 Top-3 개선/악화 질문
def show_top_cases(metric, k=3):
    delta_col = f"delta_{metric}"
    cols = [ID_COL, f"{metric}_QA", f"{metric}_REF", delta_col]

    top_improved = merged.sort_values(delta_col, ascending=False).head(k)[cols]
    top_worsened = merged.sort_values(delta_col, ascending=True).head(k)[cols]

    print(f"\n[{metric}] Top-{k} 개선 사례 (차이가 큰 순)")
    display(top_improved)

    print(f"\n[{metric}] Top-{k} 악화 사례 (차이가 작은 순)")
    display(top_worsened)

for m in METRICS:
    show_top_cases(m, k=3)
```
